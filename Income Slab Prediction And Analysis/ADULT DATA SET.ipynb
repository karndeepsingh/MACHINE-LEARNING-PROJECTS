```python
def rep(cols):
    occ = cols[0]
    s = cols[1]
    if pd.isnull(occ):
        if s == "Male":
            return "Craft-repair"
        elif s == "Female":
            return "Adm-clerical"
    return occ    
        
    
```

```python
data1["occupation"] = data1[["occupation", "sex"]].apply(rep, axis = 1)
```

```python
data_new = pd.get_dummies(data1,columns = ["income"], drop_first = True)
```

```python
data_new.head()
```

```python
data_new.columns
```

```python
from sklearn.preprocessing import LabelEncoder
lb =LabelEncoder()
```

```python
ctg = data_new[["workclass","occupation", "sex","native.country","income_>50K"]]
for i in ctg.columns[:-1]:
    t = ctg.groupby(i).mean()
    enc=pd.Series(data = lb.fit_transform(t["income_>50K"]), index = t.index)         
    ctg[i+ "_" + "encode"] = ctg[i].map(enc)
```

```python
ctg["education"] = data_new["education.num"]
```

```python
num = data_new[["fnlwgt", "capital.gain", "capital.loss", "hours.per.week", "income_>50K"]]
```

```python
num.head()
```

```python
for i in num.columns[:-1]:
    o = num.pivot_table(index = i, values = "income_>50K")
    encoded = pd.Series(o["income_>50K"], index = o.index)
    num[i + "_" + "encode"] = num[i].map(encoded)
```

```python
num = num.iloc[:, 5:]
```

```python
ctg = ctg.iloc[:, 4:]
```

```python
combined_data = pd.concat([ctg, num], axis = 1)
```

```python
combined_data
```

```python
y =combined_data.iloc[:,0]
x =combined_data.iloc[:,1:]
```

```python
sns.heatmap(combined_data.corr(), cmap = "RdYlGn", annot = True)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
import matplotlib.pyplot as plt
feat_impo = pd.Series(model.feature_importances_, index = x.columns)
feat_impo.nlargest(10).plot(kind ="barh")
```

```python
x = x.iloc[:, [1,4,5,6,7,8]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc_xtrain= sc.fit_transform(x_train)
sc_xtest = sc.fit_transform(x_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(sc_xtrain, y_train)
```

```python
y_pred = classifier.predict(sc_xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
